# **Load image in memory as PIL image and as OpenCV image**

```
# download an image from the internet and save it as 'test.png'

#!wget -O test.png  http://images.math.cnrs.fr/IMG/png/section8-image.png 
!wget -O test.png  https://cdn.openai.com/image-gpt/completions/igpt-xl-miscellaneous-33-orig.png
```

```
pimg = Pil_image.get('test.png') # load 'test.png' as a PIL image
print("Size:", pimg.size, " - Image is stored in memory as type:", type(pimg))
Pil_image.show(pimg) # show the PIL image in Colab
```

```
cvimg = Cv_image.get('test.png') # load 'test.png' as a OpenCV image (which is a numpy array of BGR colors)
print("Size:", cvimg.shape, " - Image is stored in memory as type:", type(cvimg))
Cv_image.show(cvimg) # show the OpenCV image in Colab
```

# **Super Resolution upscaling**

```
a = Image_Super_Resolution() # it can take a very long time to execute (only the first time that you use this class)
```

## Super Resolution upscaling for PIL images

```
# for reference, this is the traditional upscaling, without artificial intelligence
pimg_simple = a.simple_upscale_PIL_image(pimg,4) # upscale by 4 times in the traditional way (not using Artificial Intelligence)
print("Size:", pimg_simple.size)
Pil_image.show(pimg_simple)
```

```
pimg_super = a.super_upscale_PIL_image_by_4x(pimg) # upscale by 4 times using Artificial Intelligence
print("Size:", pimg_super.size)
Pil_image.show(pimg_super) 
# the image below should look much better than the image above
```

## Super Resolution upscaling for Open_CV images

```
# for reference, this is the traditional upscaling, without artificial intelligence
cvimg_simple = a.simple_upscale_cvimage_image(cvimg, 4.0) # upscale by 4 times in the traditional way (not using Artificial Intelligence)
print("Size:", cvimg_simple.shape)
Cv_image.show(cvimg_simple)
```

```
cvimg_super = a.super_upscale_cvimage_by_4x(cvimg)
print("Size:", cvimg_super.shape)
Cv_image.show(cvimg_super)
```
